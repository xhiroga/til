```python
def read_infections(file_name: str) -> list[int]:
    with open(file_name, "r") as f:
        line = f.readline().strip()
        return [int(word) for word in line.split(":")]
```

```python
def to_diffs(infections: list[int]) -> list[int]:
    yesterday = 0
    diffs = []
    for infection in infections:
        diff = infection - yesterday
        diffs.append(diff)
        yesterday = infection

    print(f"{diffs=}")
    return diffs
```

```python
data = [10, 15, 0]
expected = [10, 5, -15]
actual = to_diffs(data)
assert actual == expected
```

```python
# まず増加数の和が最大となる期間を全て抽出する
def find_maximum_sum_period(diffs: list[int]) -> list[tuple[int]]:
    最大の増加数の和 = 0
    最大の増加数の和の期間s = []
    現在の増加数の和 = 0
    現在の増加が開始した日 = None

    for i, diff in enumerate(diffs):
        print(f"{i=}, {diff=}")
        現在の増加数の和 += diff
        if 現在の増加が開始した日 is None:
            現在の増加が開始した日 = i

        if 現在の増加数の和 <= 0:
            現在の増加数の和 = 0
            現在の増加が開始した日 = None
            print(f"{i=}, {diff=}, {現在の増加数の和=} <= 0")
            continue

        if (
            最大の増加数の和 == 現在の増加数の和
            and 最大の増加数の和の期間s[-1][0] != 現在の増加が開始した日
        ):
            最大の増加数の和の期間s.append((現在の増加が開始した日, i))
            print(
                f"{i=}, {diff=}, {最大の増加数の和=}, {現在の増加数の和=}, {最大の増加数の和の期間s=}"
            )
        elif 最大の増加数の和 < 現在の増加数の和:
            最大の増加数の和 = 現在の増加数の和
            最大の増加数の和の期間s = [(現在の増加が開始した日, i)]
            print(
                f"{i=}, {diff=}, {最大の増加数の和=}, {現在の増加数の和=}, {最大の増加数の和の期間s=}"
            )

    return 最大の増加数の和の期間s
```

```python
# 微減する日もあるが、最終的に最大になるケース
data = [10, 20, -20, 100]
expected = [(0, 3)]
actual = find_maximum_sum_period(data)
assert actual == expected
```

```python
# 最大の増加数の和を持つ期間が複数あるケース
data = [10, 20, -50, 15, 15, -20]
expected = [(0, 1), (3, 4)]
actual = find_maximum_sum_period(data)
assert actual == expected
```

```python
# 増加数が0の期間があるケース
data = [10, 20, -50, 15, 15, 0]
expected = [(0, 1), (3, 4)]
actual = find_maximum_sum_period(data)
assert actual == expected
```

```python
def to_week_average(infections: list[int]):
    averages = []
    for i in range(3, len(infections) - 3):
        averages.append(sum(infections[i - 3 : i + 4]) / 7)
    return averages
```

```python
data = [1, 2, 1, 2, 1, 2, 1, 3, 4]
expected = [10 / 7, 12 / 7, 2]
actual = to_week_average(data)
assert actual == expected
```

```python
def calc_similarity(infection_x: list[int], infection_y: list[int]):
    sims = []
    i = len(infection_x) - len(infection_y) + 1
    for j in range(i):
        diff = 0
        for k in range(len(infection_y)):
            diff += (infection_x[k + j] - infection_y[k]) ** 2

        print(
            f"infection_x[{j}:{j+len(infection_y)}], infection_y[{0}:{len(infection_y)}], {diff=}"
        )
        sims.append(diff)
    return -1 * min(sims)
```

```python
# 類似度がマイナスになるケース
test_x = [4, 5, 6, 7]
test_y = [2, 3, 4, 5]
expected = -16
actual = calc_similarity(test_x, test_y)
assert actual == expected
```

```python
# 類似度がゼロになるケース
test_x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
test_y = [2, 3, 4, 5]
expected = 0
actual = calc_similarity(test_x, test_y)
assert actual == expected
```

```python
import sys


def question2_2():
    max_int = sys.maxsize
    min_int = -max_int - 1

    infections_list = []
    dir = "sozojyoho_2022_test_given_data/data"
    for file_name in os.listdir(dir):
        infections = read_infections(f"{dir}/{file_name}")
        infections_list.append(infections)

    max_sim = min_int
    max_sim_pairs = []
    for i, infections_i in enumerate(infections_list):
        for j, infections_j in enumerate(infections_list):
            print(f"{i=}, {j=}, {len(infections_i)=}, {len(infections_j)=}")
            if i >= j:
                print(
                    f"{i=}, {j=}, {len(infections_i)=}, {len(infections_j)=}, continue"
                )
                continue

            sim = None
            if len(infections_i) > len(infections_j):
                sim = calc_similarity(infections_i, infections_j)
            else:
                sim = calc_similarity(infections_j, infections_i)

            if max_sim == sim:
                max_sim_pairs.append((i, j))
                print(
                    f"{i=}, {j=}, {len(infections_i)=}, {len(infections_j)=}, {sim=}, append"
                )
            elif max_sim < sim:
                max_sim = sim
                max_sim_pairs = [(i, j)]
                print(
                    f"{i=}, {j=}, {len(infections_i)=}, {len(infections_j)=}, {sim=}, update"
                )

    print(f"{max_sim=}, {max_sim_pairs=}")
    return (max_sim, max_sim_pairs)


question2_2()
```

```python
def calc_params(infections: list[int]):
    n = len(infections)
    sum_i = 0
    sum_i2 = 0
    sum_xi = 0
    sum_ixi = 0
    for i, x in enumerate(infections):
        sum_i += i
        sum_i2 += i**2
        sum_xi += x
        sum_ixi += i * x

    a = (n * sum_ixi - sum_i * sum_xi) / (n * sum_i2 - sum_i**2)
    k = (sum_i2 * sum_xi - sum_ixi * sum_i) / (n * sum_i2 - sum_i**2)
    return (a, k)
```

```python
data = [2, 4, 6, 8, 10, 12]
a, k = calc_params(data)
print(f"{a=}, {k=}")
approx = [a * i + k for i, x in enumerate(data)]
print(f"{approx=}")
```

# 問2 (4) について

- 指数回帰の問題
- 誤差関数内で対数を取ることで、線形回帰同様に最小二乗法で解けるようになる
- すべてのSに対して正規方程式を計算してa,kを求め、それが最小となるSを返す。

```python
import math


def error_log(infections, k, a):
    try:
        error = 0
        for i, infection in enumerate(infections[0:31]):
            error += (math.log(k * a**i, math.e) - math.log(infection + 1, math.e)) ** 2
        return error
    except Exception:
        pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math


# mは多項式の次元を表し、この問では1となる
def resolve(xs, ts, m):
    phi = np.array([[x**k for k in range(m + 1)] for x in xs])
    tmp = np.linalg.inv(np.dot(phi.T, phi))
    log_ts = [math.log(t + 1, math.e) for t in ts]
    log_k, log_a = np.dot(np.dot(tmp, phi.T), log_ts)
    return np.exp(log_k), np.exp(log_a)
```

```python
test_x = [i for i in range(31)]
print(f"{test_x=}")
test_k, test_a = 2, 1.1


def test_fx(x, k, a):
    return k * a**x


test_y = [test_fx(x, test_k, test_a) for x in test_x]
print(f"{test_y=}")

_, axes = plt.subplots(1, 2)
axes[0].plot(test_x, test_y)

actual_k, actual_a = resolve(test_x, test_y, 1)
print(f"{actual_k=}, {actual_a=}")
test_y_hat = [test_fx(x, actual_k, actual_a) for x in test_x]
print(f"{test_y_hat=}")
axes[1].plot(test_x, test_y_hat)

actual_error = error_log(test_x, actual_k, actual_a)
print(f"{actual_error=}")
```

```python
def question2_4():
    MAX_INT = 2**63 - 1
    infections = read_infections("sozojyoho_2022_test_given_data/infections2.txt")
    infections_array = np.array(infections)
    min_error = MAX_INT
    argmin = (-1, -1, -1)
    for s in range(len(infections_array) - 31 + 1):
        days = [day for day in range(s, s + 31)]
        k, a = resolve(days, infections_array[s : s + 31], 1)
        error = error_log(infections_array[s : s + 31], k, a)
        print(f"{s=}, {k=}, {a=}, {error=}, {min_error=}, ")
        if error < min_error:
            min_error = error
            argmin = (s, k, a)

    days = [day for day in range(s, s + 31)]
    infections_hat = [k * a**day for day in days]
    _, axes = plt.subplots(1, 2)
    axes[0].plot(days, infections_array[s : s + 31])
    axes[1].plot(days, infections_hat)

    return argmin


question2_4()
```
