```python
# 勾配の大きさの観察の実験
import numpy as np
import matplotlib.pyplot as plt

N = 2   # ミニバッチサイズ
H = 3   # 隠れ状態ベクトルの次元数
T = 20  # 時系列データの長さ

dh = np.ones((N, H))
np.random.seed(3)
Wh = np.random.randn(H, H)

norm_list = []
for i in range(T):
    dh = np.dot(dh, Wh.T)
    # L2ノルム...各要素の2乗に対して平方根を求めた値
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)

plt.plot(norm_list)
plt.xlabel('time step')
plt.ylabel('norm')
plt.show()
```
