# 最長共通部分列


次のように考えて解いた。

- 総当たりで解こうとすると手に負えない（文字列 X を m 文字、文字列 Y を n 文字として、$O(2^{mn})$かかる）
- 文字列 X を固定し、文字列 Y の j 文字目までとの最長共通部分列（以降 LCS）を、文字列 Y の j+1 文字目までの LCS で利用できないか？
- ところが、次のケースで困る。

  - $X = [a, x, y, z, b, c], Y = [a, b, c, x, y, z]$
  - $X∧Y[:1] = [a]$
  - $X∧Y[:2] = [a, b]$
  - $X∧Y[:3] = [a, b, c]$
  - $X∧Y[:4] = [a, b, c]$
  - $X∧Y[:5] = [a, x, y]$ # ここで急に b, c が脱落する。
  - $X∧Y[:6] = [a, x, y, z]$

- 本当にメモ化して再帰で解けるのか？次の 3 通りを考える（LCS を ∧ で表す）
  1. $X[:i]∧Y[:j]$と$X[:i-1]∧Y[:j]$はどのような関係か？
  2. $X[:i]∧Y[:j]$と$X[:i]∧Y[:j-1]$はどのような関係か？
  3. $X[:i]∧Y[:j]$と$X[:i-1]∧Y[:j-1]$はどのような関係か？
- 1 について、$X[:i-1]∧Y[:j]$と$X_i$を引数にして、$X[:i]∧Y[:j]$の文字数を判定できるかを考える。$Y[:j]$側のまだ消費していない文字を記録していないと難しそうだ。
- 2,3 についても同様。

- 一方で、螺旋本には、$X_i \neq $Y_j$の場合は、$X_{i-1}$ と $Y_j$ の LCS あるいは $X_i$ と $Y_{j-1}$ の LCS のどちらか長い方が $X_i$ と $Y_j$ の LCS になります、とある。この考え方に慣れる。
  - $X = [a,b,c,d], Y = [a,b,x,c]$
  - ここで、$i,j = 4$として、$X_i \neq Y_j$
  - この時、$X_i$と$Y_j$の LCS は$X_i$と$Y_{j-1}$の LCS に等しい
