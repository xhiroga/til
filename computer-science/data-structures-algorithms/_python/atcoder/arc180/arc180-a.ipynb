```python
from logging import getLogger, StreamHandler, WARNING
import os

PYTHON_LOGLEVEL = os.environ.get("PYTHON_LOGLEVEL", WARNING)
logger = getLogger(__name__)
handler = StreamHandler()
handler.setLevel(PYTHON_LOGLEVEL)
logger.setLevel(PYTHON_LOGLEVEL)
logger.addHandler(handler)
logger.propagate = False
```

```python
law = 10**9 + 7


def aba_and_bab():
    _ = input()
    s = input().strip()

    stripes = []
    length = 0
    last = s[0]
    for letter in list(s):
        logger.debug(f"{letter=}")
        if letter == last:
            if length >= 3:
                stripes.append(length)
                logger.debug(f"append, {length=}")
            length = 1
        else:
            length += 1
        last = letter
    if length > 0:
        stripes.append(length)
        logger.debug(f"append, {length=}")

    variation = 1
    for stripe in stripes:
        logger.debug(f"{stripe=}, {variation=}, {(((stripe - 1) // 2 + 1))=}")
        variation = variation * ((stripe - 1) // 2 + 1) % law

    return variation
```

```python
from unittest.mock import patch

with patch("builtins.input", side_effect=["4", "ABAB"]):
    expected = 2
    actual = aba_and_bab()
    print(actual)
    assert expected == actual
```

```python
from unittest.mock import patch

with patch("builtins.input", side_effect=["17", "BBABABAABABAAAABA"]):
    expected = 18
    actual = aba_and_bab()
    print(actual)
    assert expected == actual
```
