```python
def parse(txt: str) -> list[int]:
    print(f"{txt=}")
    nums = txt.split(",")
    walls = []
    row = None
    for num in nums:
        if row is None:
            row = num
        else:
            walls.append((int(row), int(num)))
            row = None
    print(f"{walls=}")
    return walls
```

```python
# For Debug
def draw(m: int, walls: list[int]) -> str:
    # 偶数行のブランク迷路の描画
    evens = []
    for _ in range(m + 1):
        evens.append("." + "".join([" ." for _ in range(m)]))

    # 奇数行のブランク迷路の描画
    odds = []
    for _ in range(m):
        odds.append(" " + "".join(["  " for _ in range(m)]))

    # 2つの行を交互にappend
    maze = []
    while not (len(evens) == 0 and len(odds) == 0):
        if len(maze) % 2 == 0:
            maze.append(evens.pop(0))
        else:
            maze.append(odds.pop(0))

    print(f"blank mage: {m=}, {maze=}")

    for wall in walls:
        r, c = wall
        row = maze[r]
        row_list = list(row)
        if r % 2 == 0:
            row_list[c] = "-"
            maze[r] = "".join(row_list)
        else:
            row_list[c] = "|"
            maze[r] = "".join(row_list)

    print(f"mage: {m=}, {maze=}")

    return "\n".join(maze)
```

```python
expected = """
.-.-.-.
| | | |
.-.-.-.
| | | |
.-.-.-.
| | | |
.-.-.-.
""".strip()

with open("maze0.txt", "r") as f:
    line = f.readline().strip()
    walls = parse(line)
    actual = draw(3, walls)
    assert expected == actual
```

データの持ち方がいくつか考えられる。

1. 大問 1 の壁の持ち方（点・辺を問わない行列）
2. セル.上下左右をキーとする二次元ベクトル

2-2-b でセルを基準に考えるため、案 2 でデータを持つ。


```python
from collections import defaultdict

UP = 0
LEFT = 1
DOWN = 2
RIGHT = 3


def gen_cells(walls: list[int]):
    cells = defaultdict(
        lambda: defaultdict(lambda: {UP: False, LEFT: False, DOWN: False, RIGHT: False})
    )
    # NOTE: 出題にもi,jなどが利用されるため、enumerateの返り値はindexにすること。
    for index, wall in enumerate(walls):
        i = int(index // 40)
        j = index % 40
        cells[i][j][wall] = True
        if wall == UP and i != 0:
            cells[i - 1][j][wall] = True
        elif wall == LEFT and j != 0:
            cells[i][j - 1][wall] = True
        elif wall == DOWN and i != 39:
            cells[i + 1][j][wall] = True
        elif wall == RIGHT and j != 39:
            cells[i][j + 1][wall] = True

    return cells
```

```python
def parse_sequence(ss: str) -> list[int]:
    nums = [int(s) for s in ss.split(",")]
    return nums
```

```python
def has_l(cell: dict[0:bool, 1:bool, 2:bool, 3:bool]) -> bool:
    if cell[0] and cell[1] and not (cell[2]) and not (cell[3]):
        return True
    elif not (cell[0]) and cell[1] and cell[2] and not (cell[3]):
        return True
    elif not (cell[0]) and not (cell[1]) and cell[2] and cell[3]:
        return True
    elif cell[0] and not (cell[1]) and not (cell[2]) and cell[3]:
        return True
    else:
        return False
```

```python
assert has_l({UP: False, LEFT: True, DOWN: True, RIGHT: False}) is True
assert has_l({UP: True, LEFT: False, DOWN: True, RIGHT: False}) is False
assert has_l({UP: True, LEFT: True, DOWN: True, RIGHT: False}) is False
assert has_l({UP: True, LEFT: True, DOWN: True, RIGHT: True}) is False
```

```python
def count_l(cells: dict[dict[dict[int, bool]]]):
    count = 0
    for _, row in cells.items():
        for _, cell in row.items():
            if has_l(cell):
                count += 1

    return count
```

```python
assert count_l({0: {0: {UP: False, LEFT: True, DOWN: True, RIGHT: False}}}) == 1
```

```python
with open("p.txt", "r") as f:
    ss = f.readline().strip()
    walls = parse_sequence(ss)
    cells = gen_cells(walls)
    print(count_l(cells))
```
