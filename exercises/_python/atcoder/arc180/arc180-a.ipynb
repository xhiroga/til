```python
from logging import basicConfig, root, DEBUG, WARNING

basicConfig(level=DEBUG if "get_ipython" in globals() else WARNING)


law = 10**9 + 7


def aba_and_bab() -> int:
    _ = input()
    s = input().strip()

    stripes = []
    length = 0
    last = s[0]
    for letter in list(s):
        root.debug(f"{letter=}")
        if letter == last:
            if length >= 3:
                stripes.append(length)
                root.debug(f"append, {length=}")
            length = 1
        else:
            length += 1
        last = letter
    if length > 0:
        stripes.append(length)
        root.debug(f"append, {length=}")

    variation = 1
    for stripe in stripes:
        root.debug(f"{stripe=}, {variation=}, {(((stripe - 1) // 2 + 1))=}")
        variation = variation * ((stripe - 1) // 2 + 1) % law

    return variation
```

```python
from unittest.mock import patch

with patch("builtins.input", side_effect=["4", "ABAB"]):
    expected = 2
    actual = aba_and_bab()
    print(actual)
    assert expected == actual
```
