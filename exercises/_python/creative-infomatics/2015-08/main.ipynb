```python
def based4_to_decimal(based4: str) -> int:
    nums = [int(digit) for digit in based4]
    print(f"{nums=}")

    decimal = 0
    nums.reverse()
    for i, num in enumerate(nums):
        decimal += num * (4**i)

    print(f"{decimal=}")
    return decimal


expected = 27
actual = based4_to_decimal("123")
assert expected == actual, f"{expected=}, {actual=}"
```

```python
eightry_dict = {"a": 0, "b": 1, "c": 2, "d": 3, "e": 4, "f": 5, "g": 6, "h": 7, "i": 8}


def based8_to_decimal(based8: str) -> int:
    alphabets = [digit for digit in based8]
    print(f"{alphabets=}")
    alphabets.reverse()

    decimal = 0
    for i, alphabet in enumerate(alphabets):
        num = eightry_dict[alphabet]
        decimal += num * (8**i)
        print(f"{alphabet=}, {num=}, {num * (8**i)=}, {decimal=}")

    return decimal


expected = 83
actual = based8_to_decimal("bcd")
assert expected == actual, f"{expected=}, {actual=}"
```

```python
roman_dict = {
    4: "IV",
    9: "IX",
    1: "I",
    5: "V",
    40: "XL",
    90: "XC",
    10: "X",
    50: "L",
    400: "CD",
    900: "CM",
    100: "C",
    500: "D",
    1000: "M",
}


def roman_to_decimal(roman: str) -> int:
    print(f"{roman=}")
    decimal = 0
    while len(roman):
        for num, roman_digit in roman_dict.items():
            if roman.startswith(roman_digit):
                decimal += num
                roman = roman[len(roman_digit) :]
                break

    print(f"{decimal=}")
    return decimal


expected1 = 13
actual1 = roman_to_decimal("XIII")
assert expected1 == actual1, f"{expected1=}, {actual1=}"

expected2 = 207
actual2 = roman_to_decimal("CCVII")
assert expected2 == actual2, f"{expected2=}, {actual2=}"

expected3 = 1904
actual3 = roman_to_decimal("MCMIV")
assert expected3 == actual3, f"{expected3=}, {actual3=}"
```

```python
roman_dict = {
    4: "IV",
    9: "IX",
    1: "I",
    5: "V",
    40: "XL",
    90: "XC",
    10: "X",
    50: "L",
    400: "CD",
    900: "CM",
    100: "C",
    500: "D",
    1000: "M",
}


def decimal_to_roman(decimal: str) -> str:
    nums = [int(digit) for digit in decimal]
    roman = ""
    for i, num in enumerate(nums):
        power = len(nums) - i - 1
        plus = ""
        if num == 0:
            print(f"{power=}, {num=}, skip")
            continue
        elif num == 4 or num == 9:
            plus = roman_dict[num * (10**power)]
        elif num >= 5:
            plus = roman_dict[5 * (10**power)] + (
                roman_dict[1 * (10**power)] * (num - 5)
            )
        else:
            plus = roman_dict[1 * (10**power)] * num
        print(f"{power=}, {num=}, {plus=}")
        roman += plus

    print(f"{roman=}")
    return roman


expected1 = "MMMCMXCIX"
actual1 = decimal_to_roman("3999")
assert expected1 == actual1, f"{expected1=}, {actual1=}"
```

```python
def decimal_to_roman_with_list(decimal: str) -> list[str]:
    nums = [int(digit) for digit in decimal]
    roman = []
    for i, num in enumerate(nums):
        power = len(nums) - i - 1
        plus = ""
        if num == 0:
            print(f"{power=}, {num=}, skip")
            continue
        elif num == 4 or num == 9:
            plus = roman_dict[num * (10**power)]
        elif num >= 5:
            plus = roman_dict[5 * (10**power)] + (
                roman_dict[1 * (10**power)] * (num - 5)
            )
        else:
            plus = roman_dict[1 * (10**power)] * num
        print(f"{power=}, {num=}, {plus=}")
        roman.append(plus)

    print(f"{roman=}")
    return roman


expected1 = ["MMM", "CM", "XC", "IX"]
actual1 = decimal_to_roman_with_list("3999")
assert expected1 == actual1, f"{expected1=}, {actual1=}"
```

```python
extended_roman_dict = {
    4: "IV",
    9: "IX",
    40: "XL",
    45: "VL",
    49: "IL",
    90: "XC",
    95: "VC",
    99: "IC",
    400: "CD",
    450: "LD",
    490: "XD",
    495: "VD",
    499: "VD",
    900: "CM",
    950: "VM",
    990: "XM",
    995: "VM",
    991: "IM",
    1000: "M",
}


def decimal_to_roman_extended(decimal: str) -> str:
    """
    組み合わせが多いため、拡張あり全てのパターンに対して組み合わせを試す
    """

    try:
        decimal_num = int(decimal)
        min_count = 4096
        min_roman = ""
        for ex_num, ex_roman in extended_roman_dict.items():
            current_roman = ""
            diff_decimal_num = decimal_num - ex_num
            if diff_decimal_num < 0:
                continue

            diff_roman_list = decimal_to_roman_with_list(str(diff_decimal_num))
            # 並べ替え
            for i, roman in enumerate(diff_roman_list):
                roman_num = [
                    key for key, value in roman_dict.items() if value == roman
                ][0]
                if ex_num >= roman_num:
                    current_roman = "".join(
                        diff_roman_list[:i] + [ex_roman] + diff_roman_list[i:]
                    )
            if current_roman == "":
                current_roman = "".join(diff_roman_list + [ex_roman])

            print(f"{current_roman=}")

            if len(current_roman) < min_count:
                min_count = len(current_roman)
                min_roman = current_roman
    except Exception as e:
        raise e

    return min_roman


expected1 = "CIL"
actual1 = decimal_to_roman_extended(149)
print(f"{expected1=}, {actual1=}")
assert len(expected1) == len(actual1)

expected2 = "IL"
actual2 = decimal_to_roman_extended(49)
print(f"{expected2=}, {actual2=}")
assert len(expected2) == len(actual2)
```
