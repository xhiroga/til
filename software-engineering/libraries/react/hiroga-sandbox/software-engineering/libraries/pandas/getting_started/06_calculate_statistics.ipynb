```python
import pandas as pd
```

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic.head()
```

```python
titanic["Age"].mean()
```

```python
type(titanic[["Age", "Fare"]])
titanic[["Age", "Fare"]].median()
```

```python
titanic[["Age", "Fare"]].describe()
```

```python
titanic.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

```python
groupby_sex =  titanic[["Sex", "Age"]].groupby("Sex")
type(groupby_sex)
```

```python
groupby_sex.mean()
```

```python
titanic_g = titanic
titanic_g["Generation"] = titanic_g["Age"] // 10
titanic_g.groupby(["Generation", "Sex"])[["Survived", "Pclass", "Fare"]].describe()
```

```python
titanic.groupby("Sex").mean()
```

```python
series_groupby = titanic.groupby("Sex")["Age"]
type(series_groupby)
series_groupby.mean()
```

```python
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

```python
titanic["Pclass"].value_counts()
```

```python
type(titanic["Pclass"].value_counts())
```

```python
type(titanic["Pclass"].describe())
```

```python
titanic["Pclass"].count()
```

```python
titanic.groupby("Pclass")["Pclass"].count() # `count`はDataFrameとDataFrameGroupByのいずれにも使える。つまり、groupbyは内部でDataFrameに対してmapのようにメソッドを適用していると推測できる。
```

```python
titanic.groupby("Pclass")["Pclass"].value_counts() # value_counts は groupby -> count のショートカットに過ぎない。groupby後にも使うことができるが、値は1種類しかないのでIndex Nameは自明である。
```

```python
type(titanic.groupby("Pclass")["Pclass"].value_counts()) # groupbyを二重に適用しているにもかかわらずSeriesとして返される。つまり、Index Nameは複数列を持てる。
```
